```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import html5lib
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
```

```python
URL_original = 'https://voz.vn/'
```

```python
# Đường dẫn đến chromedriver (bạn tải tại https://chromedriver.chromium.org/)
options = Options()
# options.add_argument("--headless")  # Xóa dòng này nếu muốn thấy trình duyệt
driver = webdriver.Chrome(options=options)

# Truy cập Wikipedia
driver.get(URL_original)
time.sleep(5)  # Đợi trang tải xong (có thể điều chỉnh thời gian nếu cần)
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
threads = soup.select("div.node-body")

thread_links = [
    URL_original + t.select("a")[0]["href"]
    for t in threads
]

print(thread_links)
```
