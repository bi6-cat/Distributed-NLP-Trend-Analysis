```python
import re
from datetime import datetime, timedelta
from urllib.parse import urlparse

def normalize_time(raw_time: str, now=None) -> str:
    """
    Chuyển time Tinhte về format DD/MM/YYYY HH:MM
    """
    if not raw_time:
        return ""

    raw_time = raw_time.strip().lower()
    now = now or datetime.now()

    # 1. Định dạng chuẩn đã có (dd/mm/yyyy hh:mm)
    try:
        parsed = datetime.strptime(raw_time, "%d/%m/%Y %H:%M")
        return parsed.strftime("%d/%m/%Y %H:%M")
    except ValueError:
        pass

    # 2. Hôm qua
    if "hôm qua" in raw_time:
        dt = now - timedelta(days=1)
        return dt.strftime("%d/%m/%Y %H:%M")

    # 3. Dạng số + đơn vị (22 phút, 13 giờ, 2 ngày…)
    match = re.search(r"(\d+)\s*(phút|giờ|ngày|tuần|tháng)", raw_time)
    if match:
        value = int(match.group(1))
        unit = match.group(2)

        delta_map = {
            "phút": timedelta(minutes=value),
            "giờ": timedelta(hours=value),
            "ngày": timedelta(days=value),
            "tuần": timedelta(weeks=value),
            "tháng": timedelta(days=value * 30),  # gần đúng
        }

        dt = now - delta_map[unit]
        return dt.strftime("%d/%m/%Y %H:%M")

    # 4. Fallback – trả về time hiện tại
    return now.strftime("%d/%m/%Y %H:%M")
 
def get_post_id(url: str) -> str | None:
    path = urlparse(url).path.rstrip('/')
    match = re.search(r'\.(\d+)$', path)
    return match.group(1) if match else None
```

```python
import random
import time
def random_sleep(min_s=2.5, max_s=5.5):
    t = round(random.uniform(min_s, max_s), 2)
    time.sleep(t)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

all_articles = []
all_comments = []

x = 0
for i in link_list:
    if x>5: break
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)

    try:
        driver.get(i)
        random_sleep(3, 6)

        # Scroll để load full comment
        for _ in range(3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        post_id = get_post_id(i)

        # ===== ARTICLE =====
        title_elem = soup.select_one('h1')
        title = title_elem.get_text(strip=True) if title_elem else ""

        # Thử 2 cách lấy author
        author = ""
        author_elem = soup.select_one('div.jsx-1357264391.author-name a')
        if author_elem:
            author = author_elem.get_text(strip=True)
        else:
            author_elem = soup.select_one('div.jsx-1070254797.info div.author-name')
            if author_elem:
                author = author_elem.get_text(strip=True)

        # Thử lấy date từ nhiều selector (3 loại)
        date = ""
        # Loại 1: span.date
        date_elem = soup.select_one('span.date')
        if date_elem:
            raw_date = date_elem.get_text(strip=True)
            date = normalize_time(raw_date)
        else:
            # Loại 2: span.jsx-1070254797.since span
            date_elem = soup.select_one('span.jsx-1070254797.since span')
            if date_elem:
                raw_date = date_elem.get_text(strip=True)
                date = normalize_time(raw_date)
            else:
                # Loại 3: span.jsx-1357264391.date span
                date_elem = soup.select_one('span.jsx-1357264391.date span')
                if date_elem:
                    raw_date = date_elem.get_text(strip=True)
                    date = normalize_time(raw_date)

        # Thử 3 cách lấy content
        content_text = ""
        # Loại 1: div có data-author và chứa xfBody bên trong
        content_elem = soup.select_one('div[data-author] div.xfBody')
        if content_elem:
            content_text = content_elem.get_text(separator='\n', strip=True)
        else:
            # Loại 2: thread-body-wrapper chứa xfBody
            content_elem = soup.select_one('div.jsx-1070254797.thread-body-wrapper div.xfBody')
            if content_elem:
                content_text = content_elem.get_text(separator='\n', strip=True)
            else:
                # Loại 3: xfBody trực tiếp có data-author (không có wrapper)
                content_elem = soup.select_one('div.xfBody[data-author]')
                if content_elem:
                    content_text = content_elem.get_text(separator='\n', strip=True)

        all_articles.append({
            "id": post_id,
            "tac_gia": author,
            "tieu_de": title,
            "thoi_gian": date,
            "noi_dung": content_text
        })

        # ===== COMMENTS =====
        comment_blocks = soup.select('div.thread-comment')

        for c in comment_blocks:
            try:
                box = c.select_one('div[id^="post-"]')
                comment_id = box['id'].replace('post-', '') if box else None

                user_elem = c.select_one('a.author-name')
                user = user_elem.get_text(strip=True) if user_elem else "Unknown"

                time_elem = c.select_one('a.thread-comment__date span')
                raw_time = time_elem.get_text(strip=True) if time_elem else ""
                comment_time = normalize_time(raw_time)

                body = c.select_one('div.xfBody')
                if not body:
                    continue

                # remove quote
                for q in body.select('div.bbCodeQuote, blockquote'):
                    q.decompose()

                comment_content = body.get_text(separator='\n', strip=True)

                if comment_content:
                    all_comments.append({
                        "id_comment": comment_id,
                        "user": user,
                        "time": comment_time,
                        "comment": comment_content,
                        "id_post": post_id
                    })

            except Exception:
                continue

        print("DONE:", post_id)
        x += 1

    finally:
        driver.quit()
```

```python
articles_df = pd.DataFrame(all_articles)
comments_df = pd.DataFrame(all_comments)

articles_df.to_csv('articles.csv', index=False, encoding='utf-8-sig')
comments_df.to_csv('comments.csv', index=False, encoding='utf-8-sig')
```
