```python
from bs4 import BeautifulSoup
import requests

base_url = "https://vatvostudio.vn/category/tin-tuc-moi-nhat/"
page = 1

while True:
    url = f"{base_url}page/{page}/"
    r = requests.get(url)

    if r.status_code != 200:
        break

    soup = BeautifulSoup(r.text, "html.parser")

    articles = soup.select("h3.post__title a")

    if not articles:
        break

    for a in articles:
        link = a["href"]
        links.append(link)
        print(len(links))

    page += 1
```

```python
# Đường dẫn đến chromedriver (bạn tải tại https://chromedriver.chromium.org/)
options = Options()
# options.add_argument("--headless")  # Xóa dòng này nếu muốn thấy trình duyệt
driver = webdriver.Chrome(options=options)

# Truy cập Wikipedia
driver.get(i)
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup
```

```python
import re


body = soup.select_one('body')
body
id = None
if body and body.get("class"):
    m = re.search(r'postid-(\d+)', ' '.join(body["class"])) # Tìm kiếm lớp có định dạng 'postid-12345'
    if m:
        id = m.group(1)
id
```

```python
author_el = soup.select_one('div.entry-meta a[rel="author"]')
author = author_el.get_text(strip=True) if author_el else None
author
```

```python
title_el = soup.select_one('h1.entry-title')
title = title_el.get_text(strip=True) if title_el else None
title
```

```python
time_el = soup.select_one('div.entry-meta time')
time_str = time_el['datetime'] if time_el and time_el.has_attr('datetime') else None
time_format = datetime.fromisoformat(time_str)
thoi_gian = time_format.strftime("%d/%m/%Y %H:%M")
thoi_gian
```
